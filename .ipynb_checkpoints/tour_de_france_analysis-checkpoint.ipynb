```python
import pandas as pd
```

```python
df = pd.read_excel("tour_de_france.xlsx")
df.head() # first five lines of data
```

```python
df.info() #concise summary
```

```python
df.describe() #summary stats about the data
```

```python
# handling missing values
df_missing_values = df.isnull().sum()
print(df_missing_values[df_missing_values > 0])  # Display columns with missing values
```

```python
#drop rows
df.dropna(inplace=True)
```

```python
# fill in winner's avg speed
df_winner_time = df["Winner's avg speed"].fillna(0)
```

```python
# fill in winners
df_winner = df['Winner'].fillna('unknown')
```

```python
# fill the mean speed
df["Winner's avg speed"].fillna(df["Winner's avg speed"].mean())
```

```python
# fill the mean total distance
df["Total distance (km)"].fillna(df["Total distance (km)"].mean())
```

```python
# fill the mode
df["Winner's Nationality"].fillna(df["Winner's Nationality"].mode())
```

```python
# fill the median
df["Year"] = df["Year"].fillna(df["Year"].median())
```

```python
df.drop_duplicates(inplace=True)
```
